# Tensorflow Graphs

## Creating first graph

```
import tensorflow as tf
```

```
n1 = tf.constant(1) # node-1
```

```
n2 = tf.constant(2) # node-2
```

```
n3 = n1 + n2 # node-1 and node-2 yields node-3
```

```
with tf.Session() as sess:
    result = sess.run(n3)
```

```
print(result)
```

## Another simple graph

```
import numpy as np
```

```
np.random.seed(101)
tf.set_random_seed(101)

rand_a = np.random.uniform(0, 100, (5, 5))
print(rand_a)

print("\n")
rand_b = np.random.uniform(0, 100, (5, 1))
print(rand_b)
```

```
a = tf.placeholder(tf.float32)
```

```
b = tf.placeholder(tf.float32)
```

```
add_op = a + b
```

```
mul_op = a * b
```

```
with tf.Session() as sess:
    add_result = sess.run(add_op, feed_dict = {a:10, b:30})
    mul_result = sess.run(mul_op, feed_dict = {a:10, b:30})
    print(add_result)
    print("\n")
    print(mul_result)
```

```
with tf.Session() as sess:
    add_result = sess.run(add_op, feed_dict = {a:rand_a, b:rand_b})
    mul_result = sess.run(mul_op, feed_dict = {a:rand_a, b:rand_b})
    print(add_result)
    print("\n")
    print(mul_result)
```
