# Getting started with tensorflow

## Getting to know the tensorflow version

```
import tensorflow as tf
```

```
print(tf.__version__)
```

```
hello_world = tf.constant("Hello, World!")
```

```
sess = tf.Session()
```

```
print(sess.run(hello_world))
```

## Creating first simplest tensor

```
hello = tf.constant("Hello, ")
```

```
world = tf.constant("World!")
```

```
type(hello)
```

## Creating a tensorflow session

```
with tf.Session() as sess:
    result = sess.run(hello + world)
```

```
print(result)
```

## First addition operation

```
a = tf.constant(10)
b = tf.constant(20)
```

```
type(a)
```

```
a + b
```

```
a + b
```

```
with tf.Session() as sess:
    result = sess.run(a + b)
```

```
print(result)
```

## Creating first matrix

```
const = tf.constant(10)
```

```
# Creating a 4x4 matrix
fill_mat = tf.fill((4, 4), 10)
```

```
myzeros = tf.zeros((4, 4))
```

```
myones = tf.ones((4, 4))
```

```
myrand_normal = tf.random_normal((4, 4), mean = 0, stddev = 1.0)
```

```
myrand_uniform = tf.random_uniform((4, 4), minval = 0, maxval = 1)
```

```
myzeros
```

```
my_ops = [const, fill_mat, myzeros, myones, myrand_normal, myrand_uniform]
```

## Using Interactive Session for Jupyter Notebooks

```
sess = tf.InteractiveSession() # Only usable with the jupyter notebook
```

```
for op in my_ops:
    print(sess.run(op))
    print("\n")
```

## Working with matrix

```
a = tf.constant([
    [1,2],
    [3,4]
])
```

```
a.get_shape()
```

```
b = tf.constant([
    [10],
    [20]
])
```

```
b.get_shape()
```

```
result = tf.matmul(a, b)
```

```
sess.run(result)
```
