# Tensorflow Variables and Placeholders

## Creating first variable

```
import tensorflow as tf
```

```
a = tf.constant(10)
```

```
my_var = tf.Variable(initial_value = a)
```

```
print(my_var)
```

```
init = tf.global_variables_initializer()
```

```
with tf.Session() as sess:
    sess.run(init)
    result = sess.run(my_var)
```

```
print(result)
```
