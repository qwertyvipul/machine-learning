```
# Generating multiple sequences
import numpy as np

in_seq_1 = []
in_seq_2 = []
in_seq_3 = []

for i in range(1, 1001):
    in_seq_1.append(i*10)
    in_seq_2.append(i*10)
    in_seq_3.append(i*20)
    
in_seq_1 = np.array(in_seq_1)
in_seq_2 = np.array(in_seq_2)
in_seq_3 = np.array(in_seq_3)
```

```
# Convert to [rows, columns] structure
in_seq_1 = in_seq_1.reshape(in_seq_1.shape[0], 1)
in_seq_2 = in_seq_2.reshape(in_seq_2.shape[0], 1)
in_seq_3 = in_seq_3.reshape(in_seq_3.shape[0], 1)

# Horizontally stack columns
dataset = np.hstack((in_seq_1, in_seq_2, in_seq_3))
```

```
dataset
```

```
def split_sequences(sequences, n_steps):
    X, y = list(), list()
    for i in range(len(sequences)):
        end_ix = i + n_steps
        if end_ix > len(sequences) - 1:
            break
            
        seq_x, seq_y = sequences[i:end_ix, :], sequences[end_ix, :]
        X.append(seq_x)
        y.append(seq_y)
        
    return np.array(X), np.array(y)
```

```
n_steps = 3
X, y = split_sequences(dataset, n_steps)

for i in range(len(X)):
    print(X[i], y[i])
```

```
X
```

```
n_features = X.shape[2]
print(n_features)
```

```
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
```

```
model = Sequential()
model.add(LSTM(100, activation = "relu", return_sequences = True, input_shape = (n_steps, n_features)))
model.add(LSTM(100, activation = "relu"))
model.add(Dense(n_features))
```

```
model.compile(optimizer = "adam", loss = "mse")
```

```
model.fit(X, y, epochs = 400, batch_size = 10, shuffle = True, verbose = 1)
```

```
test_input = np.array([[100, 100, 200], [110, 110, 220], [120, 120, 240]])
test_input = test_input.reshape(1, n_steps, n_features)
model.predict(test_input, verbose = 0)
```

```
y_pred = model.predict(X, verbose = 1)
```

```
for i in range(len(y_pred)):
    print(y[i])
    print(y_pred[i])
    print("-------------------")
```
