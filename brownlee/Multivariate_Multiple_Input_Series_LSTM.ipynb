```
# Generating multiple sequence
import numpy as np
in_seq_1 = []
in_seq_2 = []

for i in range(1, 1001):
    in_seq_1.append(i*10 + 2)
    in_seq_2.append(i*10 + 3)
    
in_seq_1 = np.array(in_seq_1)
in_seq_2 = np.array(in_seq_2)
out_seq = np.array([in_seq_1[i] + in_seq_2[i] for i in range(len(in_seq_1))])
```

```
print(in_seq_1)
print(in_seq_2)
print(out_seq)
```

```
# Convert to [rows, columns] structure
in_seq_1 = in_seq_1.reshape(in_seq_1.shape[0], 1)
in_seq_2 = in_seq_2.reshape(in_seq_2.shape[0], 1)
out_seq = out_seq.reshape(out_seq.shape[0], 1)

# Horizontally stack columns
dataset = np.hstack((in_seq_1, in_seq_2, out_seq))
```

```
print(dataset)
```

```
def split_sequences(sequences, n_steps):
    X, y = list(), list()
    for i in range(len(sequences)):
        end_ix = i + n_steps
        if(end_ix) > len(sequences) - 1:
            break
        
        seq_x, seq_y = sequences[i:end_ix, :-1], sequences[end_ix-1, -1]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

```
n_steps = 3
X, y = split_sequences(dataset, n_steps)

for i in range(len(X)):
    print(X[i], y[i])
```

```
n_features = X.shape[2]
print(n_features)
```

```
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
```

```
model = Sequential()
model.add(LSTM(50, activation = "relu", input_shape = (n_steps, n_features)))
model.add(Dense(1))
```

```
model.compile(optimizer = "adam", loss = "mse")
```

```
model.fit(X, y, epochs = 200, verbose = 0)
```

```
test_input = np.array([[19, 21], [29, 31], [39, 41]])
test_input = test_input.reshape(1, n_steps, n_features)
model.predict(test_input, verbose = 0)
```

```
# https://keras.io/getting-started/faq/#how-can-i-save-a-keras-model
model.save("multivariate-multi-input-lstm.h5")
```
