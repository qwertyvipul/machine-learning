```
# Lets generate a sequence to act as our dataset
# In our sequence the next number will be 10 greater than the previous one

raw_seq = []
for i in range(1, 10001):
    raw_seq.append(i*10)
    
print(raw_seq)
```

```
# Split the sequence to features and label
# Each sequence has a specified number of input time steps and one output time step
import numpy as np

def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
        end_ix = i + n_steps
        if(end_ix) > len(sequence) - 1:
            break
        
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

```
split_sequence([1, 2, 3, 4, 5, 6], 2)
```

```
n_steps = 3
X, y = split_sequence(raw_seq, n_steps)

for i in range(len(X)):
    print(X[i], y[i])
```

```
# Reshape X from [samples, timesteps] to [samples, timesteps, features]
n_features = 1
X = X.reshape((X.shape[0], X.shape[1], n_features))
X
```

```
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
```

```
model = Sequential()
model.add(LSTM(50, activation = "relu", input_shape = (n_steps, n_features)))
model.add(Dense(1))
```

```
model.compile(optimizer = "adam", loss = "mse")
```

```
model.fit(X, y, epochs = 200, verbose = 0)
```

```
test_input = np.array([15, 25, 35])
test_input = test_input.reshape(1, n_steps, n_features)
model.predict(test_input, verbose = 0)
```
